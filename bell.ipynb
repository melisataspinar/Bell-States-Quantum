# BELL STATES

Here, we write the quantum circuits that produce the four Bell states.
Then we execute each circuit on a simulator for
1024 shots and plot the histogram of the probability distributions for each state. Finally, we obtain the state vector for each state and plot it on the Bloch sphere.

Let us first import the necessary libraries.

```python
import numpy as np
import qiskit as qk

from qiskit import *
from qiskit.tools.jupyter import *
from qiskit.tools.monitor import job_monitor
from qiskit.visualization import *
from qiskit.providers.aer import QasmSimulator
from qiskit.circuit.library.standard_gates import *

from ibm_quantum_widgets import *
from matplotlib import style

provider = IBMQ.load_account()

style.use("dark_background")
backend = Aer.get_backend( 'qasm_simulator' )
simulator = Aer.get_backend("aer_simulator")
statevec_sim = Aer.get_backend("statevector_simulator")
```

Let us create our quantum register and our clasical register.

```python
q_reg = qk.QuantumRegister( 2,'q' )
c_reg = qk.ClassicalRegister( 2,'c' )
```

****************************

THE FIRST BELL STATE

Now we can implement the quantum circuit that will produce the first Bell state. To achieve this, let us implement a function named bell_state_1() that returns the circuit.

```python
def bell_state_1():
    qc = QuantumCircuit( q_reg, c_reg )

    qc.h( q_reg[0] )
    qc.cx( q_reg[0], q_reg[1] )

    return qc
```

Now let us draw the circuit that produces the first Bell state.

```python
qc1 = bell_state_1()
qc1.draw()
```

Let's run this circuit on a simulator for 1024 shots, and then plot the histogram of the probability distributions.

```python
qc1.measure_all()
plot_histogram( execute( qc1, backend, shots = 1024 ).result().get_counts() )
```

Finally, let's get the state vector for this state, and plot it on the Bloch sphere.

```python
job1 = execute( qc1, statevec_sim, shots = 1024 )
plot_bloch_multivector( job1.result().get_statevector() )
```
